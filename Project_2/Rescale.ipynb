```python
import numpy as np
import cv2
import os
import sys
from pathlib import Path

count = 0
if __name__ == "__main__":
    # get directory path where the images are stored
    read_path = "./mmediting/test/LQ/"
    # get directory path where you want to save the images
    save_path = "./mmediting/test/GT/"
    #iterate through all the files in the image directory
    for _, _, image_names in os.walk(read_path):
        #iterate through all the files in the image_dir
        for image_name in image_names:
            # check for extension .png
            if '.png' in image_name:
                # get image read path(path should not contain spaces in them)
                filepath = os.path.join(read_path, image_name)
                # get image write path
                dstpath = os.path.join(save_path, image_name)
                # read the image
                img = cv2.imread(filepath)
                # do your processing
                width = int(img.shape[1] * 4)
                height = int(img.shape[0] * 4)
                dim = (width, height)
                # resize image
                resized_img = cv2.resize(img, dim, interpolation = cv2.INTER_CUBIC)

                # write the image in a different path with the same name
                cv2.imwrite(dstpath, resized_img)
                print(f'Image {image_name} finished')
                count += 1
                
    print(f'{count} images saved at {save_path}')
```
